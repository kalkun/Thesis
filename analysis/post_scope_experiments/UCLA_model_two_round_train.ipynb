# Training using two rounds

First round uses only protest labels, second round uses only violence labels. Thus, this can be seen as turning the otherwise multitask learning into a transfer learning setup.

```python
from protestDB import cursor
import time
from protestDB import models
from lib import analysis_utils as au
import pandas as pd
import os
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np
import scipy
from keras import backend as Kbackend
from keras import callbacks as Kcallbacks
from keras import optimizers as Koptimizers
from keras import losses as Klosses
from keras import models as Kmodels
from keras import backend as Kbackend
from keras import applications as Kapplications
from keras import layers as Klayers
import h5py
import itertools
from lib import transforms
```

### Loading up the data

```python
pc = cursor.ProtestCursor()
imgs = pc.getLabelledImagesAndNonProtest(source="UCLA", label_source="UCLA original")
len(imgs)
```

### Preprocessing

```python
cols = ["name", "label", "protest"]
imgs = imgs[cols]
indx_non_violence = imgs['label'].isnull()
imgs = imgs.rename(columns={"label": "violence"})
imgs
```

```python
# Apply a cut point and normalize violence scores
cutpoint = 0.43

#clips
imgs = au.clipDFColumn(imgs, 'violence', cutpoint)

# normalize
imgs = au.minMax(imgs, 'violence')

# mask values that did not had violence labels
imgs.loc[indx_non_violence, "violence"] = -1

df = imgs
df
```

### Divide the data

```python
n_folds = 5
seed = 300
```

```python
k_folds = au.getKSplitsStratified(df, n_folds, "protest", seed )
for i in k_folds:
    print("kfold has ", len(i[i["protest"] == True]), "protest images and ", len(i[i["protest"] == False]), "non protest images")
```

```python
# print first fold
k_folds[0]
```

```python
validation_df = k_folds[0].copy()
test_df = k_folds[1].copy()
train_df = k_folds[2].copy().append(k_folds[3].copy()).append(k_folds[4].copy())
print("train set has ", len(train_df) , " images. validation has ", len(validation_df), " images and test has ", len(test_df))
```

```python
batch_size = 32
```

```python
mask_value = -1; lr = 0.01; momentum = 0.9; epochs = 100; patience = 10
```

```python
hyper_params_name = "momentum-{}_batch_size-{}_n_folds-{}_n_folds_seed-{}".format(momentum, batch_size, n_folds,seed)
hyper_params_name
```

```python
train_round_1_generator = au.ResizeSequence(
    train_df,
    batch_size, 
    targets = ['protest'],
    transforms = [
        transforms.randomResizedCrop,
        transforms.randomRotation,
        transforms.randomHorizontalFlip,
        transforms.colorJitter,
        transforms.normalizeMinMax,
        transforms.lighting,
        transforms.normalizeStandardScore
    ]
)
print("%s training batches per epoch in first round" % len(train_round_1_generator))

validation_round_1_generator = au.ResizeSequence(
    validation_df,
    batch_size, 
    targets = ['protest'],
    transforms = [
        transforms.resize,
        transforms.centerCrop,
        transforms.normalizeMinMax,
        transforms.normalizeStandardScore
    ]
)

train_round_2_generator = au.ResizeSequence(
    train_df[train_df["protest"]],
    batch_size,
    targets = ["violence"],
    transforms = [
        transforms.randomResizedCrop,
        transforms.randomRotation,
        transforms.randomHorizontalFlip,
        transforms.colorJitter,
        transforms.normalizeMinMax,
        transforms.lighting,
        transforms.normalizeStandardScore,
    ]
)
print("%s training batches per epoch in second round" % len(train_round_2_generator))

validation_round_2_generator = au.ResizeSequence(
    validation_df[validation_df["protest"]],
    batch_size,
    targets = ["violence"],
    transforms = [
        transforms.resize,
        transforms.centerCrop,
        transforms.normalizeMinMax,
        transforms.normalizeStandardScore
    ]
)

test_round_1_generator = au.ResizeSequence(
    test_df,
    batch_size,
    targets = ["protest"],
    transforms = [
        transforms.resize,
        transforms.centerCrop,
        transforms.normalizeMinMax,
        transforms.normalizeStandardScore
    ]
)
print("%s test batches per epoch in first round" % len(test_round_1_generator))

test_round_2_generator = au.ResizeSequence(
    test_df[test_df["protest"]],
    batch_size, 
    targets = ['violence'],
    transforms = [
        transforms.resize,
        transforms.centerCrop,
        transforms.normalizeMinMax,
        transforms.normalizeStandardScore
    ]
)
print("%s test batches per epoch in second round" % len(train_round_2_generator))



# Modeling start
Kbackend.clear_session()

img_input = Klayers.Input(shape=(224,224,3), name='img_input')

resnet_model = Kapplications.ResNet50(include_top=False, weights = 'imagenet') (img_input)

flatten = Klayers.Flatten()(resnet_model)

protest_out = Klayers.Dense(1, activation='sigmoid', name='protest_out')(flatten)

# model = Kmodels.Model(inputs= img_input, outputs=[protest_out, violence_out])
model = Kmodels.Model(inputs = img_input, outputs=protest_out)

optimizer = Koptimizers.SGD(lr=lr, momentum=momentum, nesterov=False)
model.compile(
    optimizer=optimizer,
    loss={
        'protest_out': Klosses.binary_crossentropy,
#         'violence_out': au.buildMaskedLoss(Klosses.mean_squared_error, mask_value)
    },
    loss_weights={
        'protest_out': 1., 
#         'violence_out': 10 
    }
)
model.summary()


try:
    __file__ # Apparently, this is not defined in the notebook,
             # but it will be when we start running on the server
             # (So when things gets real)
except NameError:
    __file__ = "test"


name = au.getExperimentName(__file__, len(df), epochs, lr, hyper_params_name)
csv_results = "models_preds/%s.csv" % name
model_checkout_path = "models/%s.hdf5" % name
csv_logger_path = "logs/%s.csv" % name
tensor_board_folder = os.path.join("logs", name)

if not os.path.exists(tensor_board_folder):
    os.makedirs(tensor_board_folder)

if not os.path.exists(tensor_board_folder):
    os.makedirs("logs")

if not os.path.exists(tensor_board_folder):
    os.makedirs("models")

if not os.path.exists(tensor_board_folder):
    os.makedirs("models_preds")

print("Writing model to: \t%s" % model_checkout_path)
print("Writing log to: \t%s\n" % csv_logger_path)



# callbacks
change_lr = Kcallbacks.LearningRateScheduler(
    au.lrUpdateUCLA, 
    verbose = True
)

checkpoint = Kcallbacks.ModelCheckpoint(
    model_checkout_path,
    monitor='val_loss', 
    verbose=1,
    save_best_only=True,
    save_weights_only=False,
    mode='auto',
    period=1
)

earlystop = Kcallbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0, 
    patience=patience,
    verbose=1, 
    mode='auto'
)

tensor_board = Kcallbacks.TensorBoard(
    log_dir=tensor_board_folder,
     histogram_freq=0,
     batch_size=32, 
     write_graph=True,
     write_grads=False, 
     write_images=False, 
     embeddings_freq=0,
     embeddings_layer_names=None,
     embeddings_metadata=None
)

csv_logger = Kcallbacks.CSVLogger(
    csv_logger_path, 
    separator=',', 
    append=False
)

def train_round(model, target, train_generator, validation_generator, test_generator, round_=1):

    history = model.fit_generator(
        train_generator,
        validation_data= validation_generator,
        epochs=epochs,
        callbacks = [change_lr, checkpoint, tensor_board, csv_logger, earlystop]
    )

    print("_" * 80)
    print("Predictions:")
    print("_" * 80)
    # Preds

    #load the weights from the best model
    model.load_weights(model_checkout_path)

    preds = model.predict_generator(test_generator, verbose = 1)
    len_test = len(test_df)
    if target == 'violence':
        len_test = len(test_df[test_df["protest"]]) # Consider only protest related
    preds_target = preds[:len_test] # avoid getting more than the needed
#     preds_violence = preds[1][:len(test_df)] 
#     print("protest shape is ", preds_target.shape, " violence shape is ", preds_violence.shape)
    print("Target shape is ('%s')" % target, preds_target.shape)
    # Creates a panda data frame with the results

    test_fold_results = test_df.copy()
    if target == "violence":
        test_fold_results = test_fold_results[test_fold_results["protest"]]
    print(test_fold_results.shape)
    target_name = "preds_%s" % target
    test_fold_results[target_name] = preds_target
#     test_fold_results["pred_violence"] = preds_violence

    print("saving results")
    name = csv_results.split(".")[0] + ("_round-%s" % round_) + csv_results.split(".")[1]
    test_fold_results.to_csv(name)
    
    return model
    
pre_trained = train_round(model, "protest", train_round_1_generator, validation_round_1_generator, test_round_1_generator, 1)
print("_" * 80)
print("Done training round for protest")
print("_" * 80)

# img_input = Klayers.Input(shape=(224,224,3), name='img_input')
# resnet_model = Kapplications.ResNet50(include_top=False, weights = 'imagenet') (img_input)

# flatten = Klayers.Flatten()(resnet_model)

violence_out = Klayers.Dense(1, activation='sigmoid', name='violence_out')(flatten)
model = Kmodels.Model(inputs = img_input, outputs=violence_out)

# Load pre trained weights:
model.load_weights(model_checkout_path, by_name=True)

model.compile(
    optimizer=optimizer,
    loss={
#         'protest_out': Klosses.binary_crossentropy,
        'violence_out': Klosses.mean_squared_error
    },
    loss_weights={
#         'protest_out': 1., 
        'violence_out': 1 
    }
)
model.summary()
train_round(model, "violence", train_round_2_generator, validation_round_2_generator, test_round_2_generator, 2)
```
